<h3 style="color:#fbbc05;">2.3 Data Preprocessing: The Core of the ML Pipeline</h3>

Now that you understand your data, it's time to prepare it for modeling. This is a multi-stage process that transforms raw data into a format that ML algorithms can learn from effectively. Each step addresses specific challenges in real-world data.

**The Preprocessing Pipeline**:
1. **Data Cleaning**: Handle missing values, duplicates, inconsistencies
2. **Outlier Handling**: Identify and treat extreme values
3. **Feature Engineering**: Create new, informative features
4. **Feature Selection**: Remove irrelevant or redundant features
5. **Encoding**: Convert categorical data to numerical format
6. **Scaling**: Normalize feature ranges
7. **Balancing**: Address class imbalance (for classification)

<h4 style="color:#1a73e8;">2.3.1 Data Cleaning</h4>

Real-world data is messy. Missing values, duplicates, and inconsistencies are the norm, not the exception. Your ability to handle these issues determines your model's success.

**1. Handling Missing Values**

Missing data is ubiquitous. Pandas represents it as `NaN` (Not a Number) or `None`. Understanding *why* data is missing helps you choose the right strategy.

**Types of Missingness**:
- **MCAR (Missing Completely At Random)**: Missingness is unrelated to any variable
  - Example: A sensor randomly fails
  - Solution: Safe to delete or impute
- **MAR (Missing At Random)**: Missingness is related to observed variables but not the missing value itself
  - Example: People with higher education are more likely to report income, but income itself doesn't affect reporting
  - Solution: Can use other features to predict missing values
- **MNAR (Missing Not At Random)**: Missingness is related to the missing value itself
  - Example: People with high income are less likely to report it
  - Solution: Requires domain expertise and careful handling

**Diagnosis**:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load a dataset with missing values: Titanic
df_titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Check for missing values
print("=" * 60)
print("MISSING VALUES ANALYSIS")
print("=" * 60)
print("\nMissing Values Count:")
missing_count = df_titanic.isnull().sum()
print(missing_count[missing_count > 0])

print(f"\nMissing Values Percentage:")
missing_pct = (df_titanic.isnull().sum() / len(df_titanic)) * 100
print(missing_pct[missing_pct > 0].round(2))

# Visualize missing values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(df_titanic.isnull(), yticklabels=False, cbar=True, cmap='viridis')
plt.title('Missing Values Heatmap', fontsize=14, fontweight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Rows', fontsize=12)

# Bar plot of missing values
plt.subplot(1, 2, 2)
missing_count[missing_count > 0].plot(kind='bar', color='coral')
plt.title('Count of Missing Values by Feature', fontsize=14, fontweight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Missing Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

**Strategies for Handling Missing Values**:

**Strategy 1: Remove (Deletion)**
- **When to use**: 
  - Column has >70% missing values (likely not useful)
  - Missingness is completely random (MCAR)
  - You have plenty of data
- **How**:
  

````python
# Drop columns with >70% missing
threshold = 0.7
cols_to_drop = df_titanic.columns[df_titanic.isnull().mean() > threshold]
print(f"Columns to drop (>70% missing): {cols_to_drop.tolist()}")
df_titanic = df_titanic.drop(columns=cols_to_drop)
  
# Drop rows with any missing values (listwise deletion)
# WARNING: This can remove a lot of data!
rows_before = len(df_titanic)
df_titanic_clean = df_titanic.dropna()
rows_after = len(df_titanic_clean)
print(f"Rows removed: {rows_before - rows_after} ({((rows_before - rows_after)/rows_before)*100:.1f}%)")
  
# Drop rows where specific columns are missing
df_titanic = df_titanic.dropna(subset=['Age', 'Embarked'])
````

**Strategy 2: Impute (Fill Missing Values)**
- **For Numerical Features**:
 

````python
from sklearn.impute import SimpleImputer
  
# Strategy: Mean (for normal distributions)
imputer_mean = SimpleImputer(strategy='mean')
df_titanic['Age_mean'] = imputer_mean.fit_transform(df_titanic[['Age']])  
````

````python
# Strategy: Median (for skewed distributions or with outliers)
imputer_median = SimpleImputer(strategy='median')
df_titanic['Age'] = imputer_median.fit_transform(df_titanic[['Age']])
````

````python
# Strategy: Mode (most frequent)
imputer_mode = SimpleImputer(strategy='most_frequent')
df_titanic['Age_mode'] = imputer_mode.fit_transform(df_titanic[['Age']])
````

````python
# Strategy: Constant (fill with a specific value, e.g., 0 or -1)
imputer_constant = SimpleImputer(strategy='constant', fill_value=0)
df_titanic['Age_constant'] = imputer_constant.fit_transform(df_titanic[['Age']])  
````

````python
# Compare strategies
print("Comparison of imputation strategies:")
print(f"Original mean: {df_titanic['Age'].mean():.2f}")
print(f"Mean imputation: {df_titanic['Age_mean'].mean():.2f}")
print(f"Median imputation: {df_titanic['Age'].median():.2f}")
````

- **For Categorical Features**:
 

````python
# Most frequent (mode)
imputer_cat = SimpleImputer(strategy='most_frequent')
df_titanic['Embarked'] = imputer_cat.fit_transform(df_titanic[['Embarked']])
````

````python
# Or use pandas directly
df_titanic['Embarked'].fillna(df_titanic['Embarked'].mode()[0], inplace=True)
````

````python
# Fill with a specific category
df_titanic['Cabin'].fillna('Unknown', inplace=True)
````

**Strategy 3: Advanced Imputation**
- **KNN Imputation**: Use values from similar rows
  

````python
from sklearn.impute import KNNImputer
  
# Select numerical columns for KNN imputation
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
imputer_knn = KNNImputer(n_neighbors=5)
df_titanic[numerical_cols] = imputer_knn.fit_transform(df_titanic[numerical_cols])
````

- **Iterative Imputation**: Use other features to predict missing values
  

````python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
  
# Use Random Forest to predict missing values
imputer_iter = IterativeImputer(
      estimator=RandomForestRegressor(n_estimators=10, random_state=42),
      random_state=42,
      max_iter=10
)
df_titanic[['Age']] = imputer_iter.fit_transform(df_titanic[['Age']])
````

**2. Handling Duplicates**

Duplicate rows can bias your model, especially if they're data entry errors that create artificial patterns.

````python

# Check for duplicates
print(f"Number of duplicate rows: {df_titanic.duplicated().sum()}")

# Show duplicate rows (if any)
duplicates = df_titanic[df_titanic.duplicated(keep=False)]  # keep=False shows all duplicates
if len(duplicates) > 0:
    print("\nDuplicate rows:")
    print(duplicates)
else:
    print("No duplicate rows found.")


# Remove duplicates based on rows
df_titanic = df_titanic.drop_duplicates()

# Remove duplicates based on specific columns
# Keep first occurrence when duplicates are found
df_titanic = df_titanic.drop_duplicates(subset=['Name', 'Age'], keep='first')

# Check for near-duplicates (similar but not identical)
# Example: Same person with slightly different age due to rounding
````

<h4 style="color:#1a73e8;">2.3.2 Handling Outliers</h4>

Outliers are extreme values that differ significantly from other observations. They can distort models, especially those using distance (KNN, SVM) or variance (Linear Regression).

**Important**: Not all outliers are errors! In fraud detection, outliers *are* your signal. Always understand the context before removing outliers.

**Method 1: IQR (Interquartile Range) Method** (Non-parametric, robust)

**Steps**:
1. Calculate Q1 (25th percentile) and Q3 (75th percentile)
2. IQR = Q3 - Q1
3. Define bounds: `[Q1 - 1.5  IQR, Q3 + 1.5  IQR]`
4. Anything outside is an outlier

**Why 1.5?**: This is a standard rule that identifies approximately 0.7% of normally distributed data as outliers. You can adjust this multiplier (e.g., 3.0 for more conservative detection).

````python
# Load a realistic dataset: California Housing (built into scikit-learn)
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)

#sample the datase
df['MedHouseVal'] = housing.target  # Median house value in $100,000s



def remove_outliers_iqr(df, column, multiplier=1.5):
    """
    Remove outliers using IQR method.
    
    Parameters:
    - df: DataFrame
    - column: Column name to check for outliers
    - multiplier: IQR multiplier (default: 1.5, use 3.0 for more conservative)
    
    Returns:
    - DataFrame with outliers removed
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    
    outliers_count = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
    outliers_pct = (outliers_count / len(df)) * 100
    
    print(f"Outlier Analysis for '{column}':")
    print(f"  Q1 (25th percentile): {Q1:.2f}")
    print(f"  Q3 (75th percentile): {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower bound: {lower_bound:.2f}")
    print(f"  Upper bound: {upper_bound:.2f}")
    print(f"  Outliers found: {outliers_count} ({outliers_pct:.2f}%)")
    
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply to housing data
df_clean = remove_outliers_iqr(df, 'MedHouseVal', multiplier=1.5)

# Visualize before and after
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].boxplot(df['MedHouseVal'])
axes[0].set_title('Before Outlier Removal', fontweight='bold', fontsize=12)
axes[0].set_ylabel('Median House Value', fontsize=11)

axes[1].boxplot(df_clean['MedHouseVal'])
axes[1].set_title('After Outlier Removal (IQR Method)', fontweight='bold', fontsize=12)
axes[1].set_ylabel('Median House Value', fontsize=11)

plt.tight_layout()
plt.show()
````

**Method 2: Z-Score Method** (Parametric, assumes normality)

A Z-score tells you how many standard deviations a point is from the mean.

**Formula**: 
```
Z = (x - ) / 
```

Where:
- x = data point
-  = mean
-  = standard deviation

**Interpretation**:
- |Z| < 2: Normal (95% of data)
- 2  |Z| < 3: Possibly unusual
- |Z|  3: Outlier (99.7% threshold)

````python

from scipy import stats

def remove_outliers_zscore(df, column, threshold=3):
    """
    Remove outliers using Z-score method.
    
    Parameters:
    - df: DataFrame
    - column: Column name
    - threshold: Z-score threshold (default: 3)
    
    Returns:
    - DataFrame with outliers removed
    """
    z_scores = np.abs(stats.zscore(df[column]))
    outliers = (z_scores > threshold).sum()
    outliers_pct = (outliers / len(df)) * 100
    
    print(f"Outlier Analysis for '{column}' (Z-score method):")
    print(f"  Mean: {df[column].mean():.2f}")
    print(f"  Standard Deviation: {df[column].std():.2f}")
    print(f"  Z-score threshold: {threshold}")
    print(f"  Outliers found: {outliers} ({outliers_pct:.2f}%)")
    
    return df[z_scores <= threshold]

df_clean_zscore = remove_outliers_zscore(df, 'MedHouseVal', threshold=3)
````

**Method 3: Isolation Forest** (Advanced, for high-dimensional data)

````python

from sklearn.ensemble import IsolationForest

def detect_outliers_isolation_forest(df, columns, contamination=0.1):
    """
    Detect outliers using Isolation Forest.
    
    Parameters:
    - df: DataFrame
    - columns: List of column names to check
    - contamination: Expected proportion of outliers (default: 0.1 = 10%)
    
    Returns:
    - DataFrame with outliers removed
    """
    iso_forest = IsolationForest(contamination=contamination, random_state=42)
    outliers = iso_forest.fit_predict(df[columns])
    
    outliers_count = (outliers == -1).sum()
    print(f"Outliers detected: {outliers_count} ({(outliers_count/len(df))*100:.2f}%)")
    
    return df[outliers == 1]  # 1 = inlier, -1 = outlier

# Example usage
df_clean_iso = detect_outliers_isolation_forest(df, ['MedHouseVal', 'MedInc'], contamination=0.05)
````

> **Industry Context**: In fraud detection, you **do not remove outliers**they are your signal! Outlier handling is task-dependent. Always ask: "Is this outlier an error or a legitimate extreme value?" For example, in e-commerce, a $10,000 purchase might be an outlier, but it could be a legitimate B2B transaction.
