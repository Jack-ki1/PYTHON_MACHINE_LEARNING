<h4 style="color:#1a73e8;">2.2.2 Data Visualization: Seeing is Believing</h4>

Statistics are abstract; plots are concrete. Visualization helps you understand distributions, spot outliers, identify relationships, and communicate findings. We use three key visualizations in EDA.

**1. Histogram**: Shows the distribution of a single variable.

**What it shows**: How often different values occur (frequency distribution).

**When to use**: Understanding the shape, spread, and central tendency of a single feature.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as  np

# Load a realistic dataset: California Housing (built into scikit-learn)
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['MedHouseVal'] = housing.target  # Median house value in $100,000s


#understand the data you have
display(df.columns)
display(df.head(10))
display(df.info())
display(df.describe())

# Plot a histogram of house values
plt.figure(figsize=(12, 6))

# Create subplot for comparison
plt.subplot(1, 2, 1)
plt.hist(df['MedHouseVal'], bins=50, color='steelblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Median House Value\n(Original Scale)', fontsize=14, fontweight='bold')
plt.xlabel('Median House Value ($100,000s)', fontsize=12)
plt.ylabel('Frequency (Number of Districts)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add statistics as text
mean_val = df['MedHouseVal'].mean()
median_val = df['MedHouseVal'].median()
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='--', linewidth=2, label=f'Median: {median_val:.2f}')
plt.legend()

# Log-transformed version (for comparison)
plt.subplot(1, 2, 2)
plt.hist(np.log1p(df['MedHouseVal']), bins=50, color='coral', edgecolor='black', alpha=0.7)
plt.title('Distribution of Median House Value\n(Log-Transformed)', fontsize=14, fontweight='bold')
plt.xlabel('Log(Median House Value + 1)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

> **Figure 2.2**: The histogram reveals a **positive skew**most districts have values between $100k$300k, but a long tail extends to $500k. The log-transformed version (right) is more symmetric, which often improves model performance. This suggests a log transformation might be beneficial later.


**2. Boxplot (Box-and-Whisker Plot)**: Visualizes the five-number summary (min, Q1, median, Q3, max) and highlights outliers.

**What it shows**:
- **Box**: Interquartile Range (IQR) from Q1 (25th percentile) to Q3 (75th percentile)
- **Line in box**: Median (50th percentile)
- **Whiskers**: Extend to 1.5  IQR from Q1 and Q3
- **Dots**: Outliers (points beyond whiskers)

**When to use**: Quickly identifying outliers, comparing distributions across groups, understanding spread.

```python
plt.figure(figsize=(12, 6))

# Single boxplot
plt.subplot(1, 2, 1)
sns.boxplot(y=df['MedHouseVal'], color='lightgreen')
plt.title('Boxplot of Median House Value', fontsize=14, fontweight='bold')
plt.ylabel('Median House Value ($100,000s)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Multiple boxplots (comparing across a categorical variable)
# Create a sample categorical variable for demonstration
df['Income_Level'] = pd.cut(df['MedInc'], bins=3, labels=['Low', 'Medium', 'High'])
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Income_Level', y='MedHouseVal', palette='Set2')
plt.title('House Value by Income Level', fontsize=14, fontweight='bold')
plt.xlabel('Income Level', fontsize=12)
plt.ylabel('Median House Value ($100,000s)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate the five-number summary
Q1 = df['MedHouseVal'].quantile(0.25)
Q2 = df['MedHouseVal'].quantile(0.50)  # Median
Q3 = df['MedHouseVal'].quantile(0.75)
IQR = Q3 - Q1
min_val = df['MedHouseVal'].min()
max_val = df['MedHouseVal'].max()

print(f"\nFive-Number Summary:")
print(f"Minimum: {min_val:.2f}")
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Median (50th percentile): {Q2:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"Maximum: {max_val:.2f}")
print(f"IQR: {IQR:.2f}")
print(f"\nOutlier bounds:")
print(f"Lower bound (Q1 - 1.5IQR): {Q1 - 1.5*IQR:.2f}")
print(f"Upper bound (Q3 + 1.5IQR): {Q3 + 1.5*IQR:.2f}")
```

> **Figure 2.3**: The box shows the IQR (from Q1 to Q3). The whiskers extend to 1.5  IQR. Points beyond are **outliers** (the dots at the top). The right plot shows how house values vary across income levelsa clear relationship that will be useful for modeling. Notice how higher income levels have higher median house values and larger spreads.

**3. Pairplot**: A matrix of scatter plots showing relationships between every pair of features.

**What it shows**: 
- **Diagonal**: Distribution of each feature (histogram or KDE)
- **Off-diagonal**: Scatter plots showing relationships between feature pairs
- **Patterns**: Linear relationships, clusters, outliers

**When to use**: Understanding relationships between multiple features, identifying correlations, spotting patterns.

```python
# For speed, use a sample of 1000 points (pairplot can be slow on large datasets)
sample_df = df.sample(1000, random_state=42)

# Select key features for visualization
features_to_plot = ['MedInc', 'HouseAge', 'AveRooms', 'MedHouseVal']

# Create pairplot
sns.pairplot(
    sample_df[features_to_plot], 
    diag_kind='kde',  # Use kernel density estimation for diagonal
    plot_kws={'alpha': 0.6, 's': 20},  # Transparency and point size
    height=2.5
)
plt.suptitle('Pairwise Relationships in California Housing Data', 
             y=1.02, fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# Calculate correlation matrix
correlation_matrix = df[features_to_plot].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix.round(3))

# Visualize correlation as heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    annot=True,  # Show correlation values
    cmap='coolwarm', 
    center=0,
    fmt='.2f',  # Format: 2 decimal places
    square=True,
    linewidths=0.5,
    cbar_kws={'label': 'Correlation Coefficient'}
)
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

> **Figure 2.4**: The pairplot reveals a strong positive correlation between `MedInc` (median income) and `MedHouseVal`. This is a powerful predictive signal for our future model. Notice how the scatter plots show clear linear relationships, while the diagonal KDE plots show the distribution shape of each feature. The correlation heatmap quantifies these relationships numerically.

**Additional Useful Visualizations**:

**Scatter Plot with Regression Line**:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='MedInc', y='MedHouseVal', alpha=0.5)
sns.regplot(data=df, x='MedInc', y='MedHouseVal', scatter=False, color='red')
plt.title('House Value vs. Median Income (with Regression Line)', fontsize=14, fontweight='bold')
plt.xlabel('Median Income', fontsize=12)
plt.ylabel('Median House Value ($100,000s)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**Violin Plot** (Combines boxplot and KDE):

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Income_Level', y='MedHouseVal', palette='Set2')
plt.title('Distribution of House Values by Income Level', fontsize=14, fontweight='bold')
plt.xlabel('Income Level', fontsize=12)
plt.ylabel('Median House Value ($100,000s)', fontsize=12)
plt.tight_layout()
plt.show()
```
